Special Functions
legendre Polynomials
The angular component of the spherical Schrodiger Equation which permits non-infinite solution 

```python
from scipy.special import legendre
x = np.linspace(0, 1, 100)
plt.plot(x, legendre(6)(x))
plt.show()
```

```python
legendre(6)
```

```python
from scipy.special import jv
x = np.linspace(0,10,100)
plt.plot(x,jv(3,x))
plt.show()
```

```python
from scipy.integrate import quad
integrand = lambda x: x**2 * np.sin(x) * np.exp(-x)
integral, integral_error = quad(integrand, 0, 1)
```

```python
integral_error
```

```python
from scipy.integrate import db1quad
integrand = lambda x, y: np.sin(x+y**2)
lwr_y = lambda x: -x
upr_y = lambda x: x**2
integral, integral_error = db1quad(integrand, 0, 1, lwr_y, upr_y)
```
