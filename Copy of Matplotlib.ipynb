```
t = np.linspace(0, 0.05, 100)
v1 = np.sin(2 * np.pi * 50 * t)
v2 = np.sin(2 * np.pi * 50 * t + 2*np.pi/3)
v3 = np.sin(2 * np.pi * 50 * t + 4*np.pi/3)

plt.figure(figsize=(10, 3))
plt.plot(t, v1, '*-', label='1st phase')
plt.plot(t, v2, 'o-', label='2nd phase')
plt.plot(t, v3, 's-', label='3rd phase')

plt.xlabel('time (s)')
plt.ylabel('amplitude (V)')
plt.grid()
plt.legend()
plt.title('3 phase sine waves');
```

```
t1 = np.linspace(-0.5, 0.5, 10000)
x1 = np.sin(1/t1)

t2 = np.linspace(-10, 10, 1000)
x2 = np.exp(t2)

t3 = np.linspace(-10, 10, 1000)
x3 = np.exp(-t3)

t4 = np.linspace(-5, 10, 1000)
x4 = t4 * np.heaviside(t4, 0)

plt.figure(figsize=(12, 9))

plt.subplot(2, 2, 1)
plt.plot(t1, x1)
plt.xlabel('time (s)')
plt.ylabel('amplitude')

plt.subplot(2, 2, 2)
plt.plot(t2, x2)
plt.xlabel('time (s)')
plt.ylabel('amplitude')

plt.subplot(2, 2, 3)
plt.plot(t3, x3)
plt.xlabel('time (s)')
plt.ylabel('amplitude')

plt.subplot(2, 2, 4)
plt.plot(t4, x4, t4, 2*x4)
plt.xlabel('time (s)')
plt.ylabel('amplitude')

plt.suptitle('Exmamples of signals that do not have Fourier transforms');
```

```
X, Y = np.mgrid[-2*np.pi : 2*np.pi : 30j, -2*np.pi : 2*np.pi : 30j]
U = Y
V = X

speed = np.sqrt(U**2 + V**2)
col = speed / np.amax(speed)

fig, ax = plt.subplots()
stream = ax.contourf(X,Y, speed, 1000, color=col, antialiased=True, 
            ls=None, alpha=0.5, cmap=plt.cm.viridis)
fig.colorbar(stream)

ax.quiver(X, Y, U, V)

plt.title('Vector field')
plt.grid()
```
